```python
import pandas as pd
```

```python
df_2015_2016_table = pd.read_html('https://fbref.com/en/comps/12/2015-2016/2015-2016-La-Liga-Stats',attrs = {"id" :"results2015-2016121_overall"})[0]
```

```python
df_2016_2017_table = pd.read_html('https://fbref.com/en/comps/12/2016-2017/2016-2017-La-Liga-Stats', attrs = {"id" : "results2016-2017121_overall"})[0]
```

```python
df_2017_2018_table = pd.read_html('https://fbref.com/en/comps/12/2017-2018/2017-2018-La-Liga-Stats', attrs={"id" : "results2017-2018121_overall"})[0]
```

```python
df_2018_2019_table = pd.read_html('https://fbref.com/en/comps/12/2018-2019/2018-2019-La-Liga-Stats', attrs={"id" : "results2018-2019121_overall"})[0]
```

```python
df_2019_2020_table = pd.read_html('https://fbref.com/en/comps/12/2019-2020/2019-2020-La-Liga-Stats', attrs = {"id" : "results2019-2020121_overall"})[0]
```

```python
df_2020_2021_table = pd.read_html('https://fbref.com/en/comps/12/2020-2021/2020-2021-La-Liga-Stats', attrs = {"id" : "results2020-2021121_overall"})[0]
```

```python
df_2021_2022_table = pd.read_html('https://fbref.com/en/comps/12/2021-2022/2021-2022-La-Liga-Stats', attrs = {"id" : "results2021-2022121_overall"})[0]
```

```python
df_2022_2023_table = pd.read_html('https://fbref.com/en/comps/12/2022-2023/2022-2023-La-Liga-Stats', attrs = {"id" : "results2022-2023121_overall"})[0]
```

```python
df_2023_2024_table = pd.read_html('https://fbref.com/en/comps/12/2023-2024/2023-2024-La-Liga-Stats', attrs = {"id" : "results2023-2024121_overall"})[0]
```

```python
df_2024_2025_table = pd.read_html('https://fbref.com/en/comps/12/2024-2025/2024-2025-La-Liga-Stats', attrs = {"id" : "results2024-2025121_overall"})[0]
```

```python
# List of data_frames
data_frames = [df_2015_2016_table,df_2016_2017_table,df_2017_2018_table,df_2018_2019_table,df_2019_2020_table,df_2020_2021_table,df_2021_2022_table,df_2022_2023_table,df_2023_2024_table,
               df_2024_2025_table]

# Standardise column names
for df in data_frames:
    df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")

# Define expected columns (ensuring missing xG-related ones are included)
expected_columns = ["rk", "squad", "mp", "w", "d", "l", "gf", "ga", "gd", "pts", "pts/mp", "attendance", "top_team_scorer", "goalkeeper", "notes", "xg", "xga", "xgd", "xgd/90" 
                    #Expected goals columns missing in older data 
                    ]

# Add a season column to track the year
years = list(range(2015, 2025))
for df, year in zip(data_frames, years):
    df["season"] = year

# Concatenate all seasons into one Pandas Data_Frame
full_data = pd.concat(data_frames, ignore_index=True)
```
